## Task - 5

### Objective: To analyze traffic accident data to identify patterns related to road conditions,weather and time of day. Visualize accident hotspots and contributing factors

### Dataset used:  Road Traffic Accidents

#### Source: Kaggle

### Description: 
##### The data set has been prepared from manual records of road traffic accident of the year 2017-20. All the sensitive information have been excluded during data encoding and finally it has 32 features and 12316 instances of the accident.

```python
#importing the libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

# Loading the dataset
df = pd.read_csv(r'C:\Users\tessa\OneDrive\Documents\MDS\Datasets\cleaned.csv')  # Replace 'accidents.csv' with your actual file name

# Data Cleaning
df.dropna(inplace=True)

# Analyzing Patterns
#Age Band of Driver
plt.figure(figsize=(12, 6))
sns.countplot(x='Age_band_of_driver', data=df)
plt.title('Accident Counts by Age Band of Driver')
plt.xticks(rotation=45)
plt.show()

#Sex of Driver
plt.figure(figsize=(8, 5))
sns.countplot(x='Sex_of_driver', data=df)
plt.title('Accident Counts by Sex of Driver')
plt.show()

#Educational Level
plt.figure(figsize=(10, 6))
sns.countplot(x='Educational_level', data=df)
plt.title('Accident Counts by Educational Level')
plt.xticks(rotation=45)
plt.show()

#Vehicle Driver Relation
plt.figure(figsize=(12, 6))
sns.countplot(x='Vehicle_driver_relation', data=df)
plt.title('Accident Counts by Vehicle Driver Relation')
plt.xticks(rotation=45)
plt.show()

#Driving Experience
plt.figure(figsize=(12, 6))
sns.countplot(x='Driving_experience', data=df)
plt.title('Accident Counts by Driving Experience')
plt.xticks(rotation=45)
plt.show()

#Road Surface Type
plt.figure(figsize=(10, 6))
sns.countplot(x='Road_surface_type', data=df)
plt.title('Accident Counts by Road Surface Type')
plt.xticks(rotation=45)
plt.show()

#Weather Conditions
plt.figure(figsize=(10, 6))
sns.countplot(x='Weather_conditions', data=df)
plt.title('Accident Counts by Weather Conditions')
plt.xticks(rotation=45)
plt.show()

#Cause of Accident
plt.figure(figsize=(12, 6))
sns.countplot(x='Cause_of_accident', data=df)
plt.title('Accident Counts by Cause of Accident')
plt.xticks(rotation=45)
plt.show()

```

### Inferences:
##### 1) Most accidents occur for individuals between 18-30 years of age.
##### 2) Fresh high school graduates contribute most to accidents.
##### 3)Employees are most prone to accidents due to extreme curfew times.
##### 4) Most accidents take place on asphalt roads.
##### 5) Ignorance of pedestrians is the often the reason for an accident
